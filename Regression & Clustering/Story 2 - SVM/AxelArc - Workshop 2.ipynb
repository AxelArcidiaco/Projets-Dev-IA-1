```python
# Importation des librairies Python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importation du dataset
dataset = pd.read_csv('./Data/Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values.astype(float)
y = dataset.iloc[:, 2:3].values.astype(float)

# 3 Feature Scaling
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
    
# Ajustement du modèle à nos données
regressor = SVR(kernel='rbf')
regressor.fit(X, y)

# Prediction du modèle
y_pred = regressor.predict(np.array([6.5]).reshape(1, 1))
print("Prediction : {}".format(y_pred))
```

```python
# Visualization des résultats de la régression vectorielle de support
plt.scatter(X, y, color = 'magenta')
plt.plot(X, regressor.predict(X), color = 'green')
plt.title('Truth or Bluff (Support Vector Regression Model)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
# Visualization des résultats de la régression vectorielle de support après le Feature Scaling
plt.scatter(X, y, color = 'magenta')
plt.plot(X, regressor.predict(X), color = 'green')
plt.title('Truth or Bluff (Support Vector Regression Model)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
#5 Predicting a new result
y_pred = sc_y.inverse_transform((regressor.predict (sc_X.transform(np.array([6.5]).reshape(-1, 1)))))
y_pred
```

```python
# Visualising the Regression results (for higher resolution and #smoother curve)
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Support Vector Regression Model(High Resolution))')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
